## Import Libraries & Set Up

```python
# import library
import warnings
import re
import os
import sys
import numpy as np
import pandas as pd
from datetime import datetime, timedelta

import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
from prophet import Prophet
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from sklearn.model_selection import TimeSeriesSplit
from sklearn.preprocessing import StandardScaler
from xgboost import XGBRegressor

plt.style.use("seaborn-v0_8-whitegrid")
pd.set_option("display.float_format", lambda x: f"{x:,.6f}")

SEED = 42
RANDOM_STATE = 42
warnings.filterwarnings('ignore')
np.random.seed(SEED)

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (12,4)
```

```python
# set up
START, INTERVAL, H = "2020-01-01", "1d", 7
IHSG, USDIDR = "^JKSE", "USDIDR=X"
BANKS = {"BBCA.JK":"BBCA","BBRI.JK":"BBRI","BBNI.JK":"BBNI","BMRI.JK":"BMRI"}

XGB_PARAMS = dict(
    n_estimators=800, max_depth=6, learning_rate=0.03,
    subsample=0.9, colsample_bytree=0.9, reg_lambda=1.0,
    random_state=42, n_jobs=-1
)
```

## Load data + exogenous

```python
def dl(tkr):
    d = yf.download(tkr, start=START, interval=INTERVAL, progress=False)
    return d.reset_index()[["Date","Close"]].rename(columns={"Date":"ds","Close":"y"}).sort_values("ds")

def build_features(df_price, ihsg, usdx):
    df = df_price.merge(ihsg, on="ds", how="inner").merge(usdx, on="ds", how="inner")
    df["ds"] = pd.to_datetime(df["ds"]).dt.tz_localize(None)
    df = df.rename(columns={"y_x":"y","y_y":"ihsg","y":"usdidr"}) if "y_x" in df.columns else df

    # exogenous returns (z-score)
    df["ihsg_ret1"] = df["ihsg"].pct_change()
    df["usdidr_ret1"] = df["usdidr"].pct_change()
    df["ihsg_ret1_z"] = (df["ihsg_ret1"]-df["ihsg_ret1"].mean())/(df["ihsg_ret1"].std()+1e-12)
    df["usdidr_ret1_z"] = (df["usdidr_ret1"]-df["usdidr_ret1"].mean())/(df["usdidr_ret1"].std()+1e-12)

    # target return & log-return (t+1)
    df["ret1"] = df["y"].pct_change()
    df["logret"] = np.log1p(df["ret1"])
    df["y_next_logret"] = df["logret"].shift(-1)  # target

    # teknikal ringkas
    for k in [1,2,3,5,7,10,14,20]:
        df[f"lag_{k}"] = df["y"].shift(k)
        df[f"ret_lag_{k}"] = df["ret1"].shift(k)
    for w in [5,10,20]:
        df[f"roll_mean_{w}"] = df["y"].rolling(w).mean()
        df[f"roll_std_{w}"]  = df["y"].rolling(w).std()

    # RSI14
    dlt = df["y"].diff(); up, down = dlt.clip(lower=0), -dlt.clip(upper=0)
    rs = up.ewm(alpha=1/14, adjust=False).mean()/(down.ewm(alpha=1/14, adjust=False).mean()+1e-12)
    df["rsi14"] = 100 - (100/(1+rs))

    # MACD & BB width
    ema12 = df["y"].ewm(span=12, adjust=False).mean()
    ema26 = df["y"].ewm(span=26, adjust=False).mean()
    macd = ema12-ema26; sig = macd.ewm(span=9, adjust=False).mean()
    df["macd"], df["macds"], df["macdh"] = macd, sig, macd - sig
    ma20, sd20 = df["y"].rolling(20).mean(), df["y"].rolling(20).std()
    df["bb_width"] = (2*sd20)/(ma20+1e-12)

    df = df.dropna().reset_index(drop=True)

    feat_cols = [
        "lag_1","lag_2","lag_3","lag_5","lag_7","lag_10","lag_14","lag_20",
        "ret_lag_1","ret_lag_2","ret_lag_3","ret_lag_5","ret_lag_7","ret_lag_10","ret_lag_14","ret_lag_20",
        "roll_mean_5","roll_mean_10","roll_mean_20","roll_std_5","roll_std_10","roll_std_20",
        "rsi14","macd","macds","macdh","bb_width","ihsg_ret1_z","usdidr_ret1_z"
    ]
    return df, feat_cols
```

```python
ihsg = dl(IHSG).rename(columns={"y":"ihsg"}); usdx = dl(USDIDR).rename(columns={"y":"usdidr"})
bbca = dl("BBCA.JK")
df_bbca, FEATS = build_features(bbca, ihsg, usdx)
```

## XGboost Model

```python
# train test split
N = len(df_bbca)
tr_end = int(0.70*N)
va_end = int(0.85*N)

train = df_bbca.iloc[:tr_end].copy()
valid = df_bbca.iloc[tr_end:va_end].copy()
test  = df_bbca.iloc[va_end:].copy()

X_tr, y_tr = train[FEATS].values, train["y_next_logret"].values
X_va, y_va = valid[FEATS].values, valid["y_next_logret"].values
X_te, y_te = test[FEATS].values,  test["y_next_logret"].values
```

```python
# build model
model = XGBRegressor(**XGB_PARAMS)

# fit
model.fit(X_tr, y_tr, eval_set=[(X_va, y_va)], verbose=False)
```

```python
# evaluate
logr_hat = model.predict(X_te)

ev = pd.DataFrame({
    "ds_t":        test["ds"].to_numpy().ravel(),           # waktu t
    "y_t":         test["y"].to_numpy().ravel(),            # harga t
    "ds_t1":       test["ds"].shift(-1).to_numpy().ravel(), # waktu t+1
    "y_next_true": test["y"].shift(-1).to_numpy().ravel(),  # harga t+1 aktual
    "logr_hat":    np.array(logr_hat).ravel(),              # pred log-return
})

ev = ev.dropna(subset=["y_next_true"]).reset_index(drop=True)

# rekonstruksi harga prediksi
ev["y_next_pred"] = ev["y_t"] * np.exp(ev["logr_hat"])

rmse = mean_squared_error(ev["y_next_true"], ev["y_next_pred"])
mae  = mean_absolute_error(ev["y_next_true"], ev["y_next_pred"])
r2   = r2_score(ev["y_next_true"], ev["y_next_pred"])
print(f"[TEST] XGB Price RMSE={rmse:.4f} | MAE={mae:.4f} | R²={r2:.4f}")

# plot
plt.figure(figsize=(12,5))
plt.plot(ev["ds_t1"], ev["y_next_true"], label="Actual (t+1)")
plt.plot(ev["ds_t1"], ev["y_next_pred"], "--", label="XGB (t+1)")
plt.title("Test Forecast — XGBoost (one-step ahead)")
plt.legend()
plt.show()
```

```python
def last_row_features_from_prices(prices: np.ndarray):
    # pastikan selalu 1D
    prices = np.asarray(prices).reshape(-1).astype(float)

    s = pd.Series(prices)
    ret1 = s.pct_change()
    feat = pd.DataFrame({"y": s, "ret1": ret1})

    for k in [1,2,3,5,7,10,14,20]:
        feat[f"lag_{k}"] = s.shift(k)
        feat[f"ret_lag_{k}"] = ret1.shift(k)
    for w in [5,10,20]:
        feat[f"roll_mean_{w}"] = s.rolling(w).mean()
        feat[f"roll_std_{w}"] = s.rolling(w).std()

    dlt = s.diff(); up = dlt.clip(lower=0); down = -dlt.clip(upper=0)
    rs = up.ewm(alpha=1/14, adjust=False).mean() / (down.ewm(alpha=1/14, adjust=False).mean()+1e-12)
    feat["rsi14"] = 100 - (100/(1+rs))

    ema12 = s.ewm(span=12, adjust=False).mean(); ema26 = s.ewm(span=26, adjust=False).mean()
    macd = ema12 - ema26; sig = macd.ewm(span=9, adjust=False).mean()
    feat["macd"], feat["macds"], feat["macdh"] = macd, sig, macd - sig

    ma20, sd20 = s.rolling(20).mean(), s.rolling(20).std()
    feat["bb_width"] = (2*sd20)/(ma20+1e-12)

    feat["ihsg_ret1_z"] = 0.0
    feat["usdidr_ret1_z"] = 0.0

    feat = feat.dropna().iloc[[-1]]
    return feat[FEATS].to_numpy(dtype=float)  # (1, n_feat)

def forecast_15d_recursive(model, df_feat):
    hist_prices = df_feat["y"].values.astype(float)
    future_prices = []
    for _ in range(H):
        x_last = last_row_features_from_prices(hist_prices[-200:])  # tail buffer
        logret_hat = float(model.predict(x_last)[0])
        logret_hat = np.clip(logret_hat, -0.06, 0.06)               # safety clip ±6%/hari
        next_price = float(hist_prices[-1]) * float(np.exp(logret_hat))
        future_prices.append(next_price)
        hist_prices = np.append(hist_prices, next_price)
    future_prices = np.asarray(future_prices, dtype=float)
    future_dates  = pd.date_range(start=df_feat["ds"].iloc[-1] + pd.Timedelta(days=1),
                                  periods=len(future_prices), freq="D")
    return future_dates, future_prices

fut_dates, fut_prices = forecast_15d_recursive(model, df_bbca)

plt.figure(figsize=(12,5))
plt.plot(df_bbca["ds"], df_bbca["y"], label="Historical")
plt.plot(fut_dates, fut_prices, "o--", label="XGB Forecast 7D")
plt.title("XGBoost — Forecast BBCCA 7 Hari ke Depan")
plt.legend()
plt.show()
```

## Ranking untuk 4 Bank (Bar & Bubble)

```python
def train_xgb_for_ticker(tkr):
    base = dl(tkr)
    df_t, feats = build_features(base, ihsg, usdx)  # FEATS sama urutan
    N = len(df_t)
    tr_end, va_end = int(0.70*N), int(0.85*N)
    train, valid = df_t.iloc[:tr_end], df_t.iloc[tr_end:va_end]
    X_tr, y_tr = train[feats].values, train["y_next_logret"].values
    X_va, y_va = valid[feats].values, valid["y_next_logret"].values
    m = XGBRegressor(**XGB_PARAMS)
    m.fit(X_tr, y_tr, eval_set=[(X_va, y_va)], verbose=False)
    return df_t, m, feats

def expected_return_15d(df_t, m, feats):
    # gunakan fungsi forecast yang sama
    def last_row_feat(prices):
        return last_row_features_from_prices(prices)
    hist_prices = df_t["y"].values.astype(float)
    path = []
    for _ in range(H):
        x_last = last_row_feat(hist_prices[-200:])
        lr = float(m.predict(x_last)[0])
        lr = np.clip(lr, -0.06, 0.06)
        nxt = float(hist_prices[-1]) * float(np.exp(lr))
        path.append(nxt)
        hist_prices = np.append(hist_prices, nxt)
    last_close = float(df_t["y"].iloc[-1])
    end_price = float(path[-1])
    return last_close, end_price, (end_price/last_close)-1.0

# BBCA: pakai model yg sudah dilatih di atas
bbca_last, bbca_end, bbca_ret = expected_return_15d(df_bbca, model, FEATS)

rows = [{"bank":"BBCA","last_close":bbca_last,"yhat_15d":bbca_end,"exp_return_15d":bbca_ret}]
for tkr, short in BANKS.items():
    if short=="BBCA": continue
    dft, m, feats = train_xgb_for_ticker(tkr)
    lc, ep, r15 = expected_return_15d(dft, m, feats)
    rows.append({"bank":short,"last_close":lc,"yhat_15d":ep,"exp_return_15d":r15})

rank_df = pd.DataFrame(rows).sort_values("exp_return_15d", ascending=False).reset_index(drop=True)
rank_df
```

```python
p = rank_df.copy()
p["return_pct"] = p["exp_return_15d"]*100.0

fig, axes = plt.subplots(1, 2, figsize=(16, 5)) # 1 row, 2 columns

# Bar plot
sns.barplot(data=p, x="bank", y="return_pct", ax=axes[0], order=p["bank"], palette="viridis")
axes[0].set_title("Expected Return 7H — XGBoost (4 Bank)")
axes[0].set_ylabel("Return 7H (%)")
for i,v in enumerate(p["return_pct"]):
    axes[0].text(i, v + (0.2 if v>=0 else -0.2), f"{v:.2f}%", ha="center",
            va="bottom" if v>=0 else "top", fontsize=9)

# Bubble plot
sizes=(p["exp_return_15d"].abs()*1500)+80
axes[1].scatter(p["last_close"], p["yhat_15d"], s=sizes, alpha=0.6, color="purple")
for _,r in p.iterrows():
    axes[1].annotate(r["bank"], (r["last_close"], r["yhat_15d"]), textcoords="offset points", xytext=(6,6))
mn=min(p["last_close"].min(), p["yhat_15d"].min())
mx=max(p["last_close"].max(), p["yhat_15d"].max())
axes[1].plot([mn,mx],[mn,mx],'--',lw=1, color="gray")
axes[1].set_xlabel("Last Close")
axes[1].set_ylabel("Forecasted Price (t+7d)")
axes[1].set_title("Bubble: Last vs Forecasted (size ~ |Return 7H|)")

plt.tight_layout() # Adjust layout to prevent overlapping
plt.show()
```

```python
# Forecast helper untuk semua bank
def forecast_for_ticker(tkr, horizon=7):
    base = dl(tkr)
    df_t, feats = build_features(base, ihsg, usdx)

    # split train/val/test → latih model
    N = len(df_t); tr_end, va_end = int(0.70*N), int(0.85*N)
    train, valid = df_t.iloc[:tr_end], df_t.iloc[tr_end:va_end]
    X_tr, y_tr = train[feats].values, train["y_next_logret"].values
    X_va, y_va = valid[feats].values, valid["y_next_logret"].values

    m = XGBRegressor(**XGB_PARAMS)
    m.fit(X_tr, y_tr, eval_set=[(X_va, y_va)], verbose=False)

    # recursive forecast H step
    hist_prices = df_t["y"].values.astype(float)
    fut_prices = []
    for _ in range(horizon):
        x_last = last_row_features_from_prices(hist_prices[-200:])
        lr = float(m.predict(x_last)[0]); lr = np.clip(lr, -0.06, 0.06)
        nxt = float(hist_prices[-1]) * np.exp(lr)
        fut_prices.append(nxt)
        hist_prices = np.append(hist_prices, nxt)

    fut_dates = pd.date_range(start=df_t["ds"].iloc[-1] + pd.Timedelta(days=1),
                              periods=len(fut_prices), freq="D")
    return df_t, fut_dates, np.asarray(fut_prices, dtype=float)


# Jalankan untuk 4 bank
results = {}
for tkr, short in BANKS.items():
    df_t, fut_dates, fut_prices = forecast_for_ticker(tkr, horizon=7)
    results[short] = {"hist": df_t, "dates": fut_dates, "pred": fut_prices}
```

```python
# Plot subplot 2x2 untuk 4 bank
fig, axes = plt.subplots(2, 2, figsize=(14, 8), sharex=False, sharey=False)
axes = axes.flatten()

for ax, (short, res) in zip(axes, results.items()):
    ax.plot(res["hist"]["ds"], res["hist"]["y"], label="Historical")
    ax.plot(res["dates"], res["pred"], "o--", label="Forecast 7D")
    ax.set_title(f"{short} — Forecast 7 Hari")
    ax.legend()

plt.suptitle("XGBoost — Forecast 7 Hari ke Depan untuk 4 Bank", fontsize=14, y=1.02)
plt.tight_layout()
plt.show()
```
